```python
# แยก feature กับ target
```

```python
df_X = df.drop(['Attrition'], axis=1)
df_X
```

```python
df_y = df['Attrition']
df_y
```

```python
# ลองแปลง feature ทั้งหมดด้วย one hot encode
```

```python
df_X = pd.get_dummies(df_X)
df_X
```

```python
df_X.keys()
```

```python
# ลองแปลง target เป็นตัวเลข
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df_y.to_list()
```

```python
le = LabelEncoder()
y = le.fit_transform(df_y.to_list())
print("shape : ",y.shape)
y
```

```python
le.classes_
```

```python
# แปลงfeatureเป็น numpy array 
```

```python
X = df_X.to_numpy()
print("shape : ",X.shape)
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=True, test_size = 0.2)
```

```python
print("X train : ", X_train.shape, "\ty train : ", y_train.shape)
print("X test : ", X_test.shape, "\ty test : ", y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=10, n_jobs=-1)
```

```python
rfc.fit(X_train, y_train)
```

```python
rfc.score(X_test, y_test)
```

```python
import sklearn
import seaborn as sns
```

```python
models = {
    'dicistiontree':sklearn.tree.DecisionTreeClassifier(),
    'logistic':sklearn.linear_model.LogisticRegression(),
    'randomforest':sklearn.ensemble.RandomForestClassifier(),
    'svm':sklearn.svm.SVC(),
    'ridge':sklearn.linear_model.RidgeClassifier(),
    'sgd':sklearn.linear_model.SGDClassifier(),
    'knn':sklearn.neighbors.KNeighborsClassifier(),
    'GradientBoosting':sklearn.ensemble.GradientBoostingClassifier(),
}
```

```python
for model_name, model in models.items():
    print(model_name)
    model.fit(X_train, y_train)
```

```python
scores = { model_name:model.score(X_test, y_test) for model_name, model in models.items() }
scores
```

```python
df_score = pd.DataFrame({'model_name':[x for x in scores], 'score':[scores[x] for x in scores]})
df_score = df_score.sort_values(by='score', ascending=False)
df_score
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='model_name', y='score', data=df_score)
plt.grid()
```
